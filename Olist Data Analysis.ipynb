## 4: What is the average order value (AOV) on Olist, and how does this vary by product category or payment method?

```python
data
```

## Group by payment type and get to the sum of total revenue for each payment type

```python
revenue_by_payment_type=data[['payment_type','payment_value']].groupby('payment_type').sum()
revenue_by_payment_type.rename(columns={'payment_value':'total_revenue'},inplace=True)
revenue_by_payment_type
```

## Group by payment type and get the number of time a purchase has been made with that payment type

```python
no_of_purchase_by_payment_type=data[['payment_type','payment_value']].groupby('payment_type').count()
no_of_purchase_by_payment_type.rename(columns={'payment_value':'number_of_purchases'},inplace=True)
no_of_purchase_by_payment_type
```

## Create a variable average_order_value thats stores the average order value of each payment type, to do this, we merge the previous two groupings we made and divide the total_revenue by number of purchases

```python
average_order_value=revenue_by_payment_type.merge(no_of_purchase_by_payment_type, on='payment_type')
average_order_value['aov']=round(average_order_value['total_revenue']/average_order_value['number_of_purchases'],2)
average_order_value.reset_index(inplace=True)
```

## Let's see the average_order_value dataframe

```python
average_order_value
```

## Drop row 3 [row with payment type 'not_defined']

```python
average_order_value=average_order_value.drop(3)
average_order_value
```

## Let's view it in a bar graph

```python
plt.bar(average_order_value['payment_type'],average_order_value['aov'])
plt.title('Average Order Value for Payment Type')
plt.xlabel('Payment Type')
plt.ylabel('Average Order Value (AOV)')
plt.show()
```

## A dictionary that maps abbreviation of the states to their name

```python
state_conversion={'AC':'Acre','AL':'Alagoas','AP':'Amapa','AM':'Amazonas','BA':'Bahia','CE':'Ceara',
                 'DF':'Distrito Federal','ES':'Espirito Santo','GO':'Goias','MA':'Maranhao','MT':'Mato Grosso',
                 'PE':'Pernambuco','MS':'Mato Grosso do Sul','MG':'Minas Gerais','PA':'para','PR':'parana','PB':'Paraiba',
                 'PI':'Piaui','RR':'Roraima','RO':'Rondonia','RJ':'Rio de Janeiro','RN':'Rio Grande do Norte',
                 'RS':'Rio Grande do Sul','SC':'Santa Catarina','SP':'Sao Paulo','SE':'Sergipe','TO':'Tocantins'}
```

## change the abbreviation to the normal name in customer state

```python
def convert_state(state):
    return state_conversion[state]
customers_df['customer_state']=customers_df['customer_state'].apply(convert_state)
```

### This contain the necessary information

## Get amount of customer per city

```python
customers_per_city=customers_df[['customer_id','customer_state']].groupby('customer_state').count()\
.sort_values('customer_id',ascending=False).reset_index()
customers_per_state=customers_per_city.rename({'customer_state':'state','customer_id':'no. of customers'}, axis=1)
customers_per_state
```

## Make a plot for the top 10 states with highest customer density

```python
plt.barh(customers_per_state.state[:10],customers_per_state['no. of customers'][:10])
plt.title('Top 10 States With Highest Customer Density')
plt.xlabel('State')
plt.ylabel('No. of customer')
plt.show()
```

## Merge data with customer_df

```python
data=data.merge(customers_df, on='customer_id')
data
```

## Check for missing values after merging

```python
data.info()
```

## Get number of total purchase for each customer in each state 

```python
customer_state_details=data[['customer_state','customer_id','order_id']].groupby(['customer_state','customer_id']).count()
customer_state_details
```

## Calculate the retention rate for each city

```python
retention_rate={}
for state in customer_state_details.index.unique('customer_state'):
    retained=(customer_state_details.loc[state]>1)['order_id'].sum()
    total=len(customer_state_details.loc[state])
    
    retention_rate[state]=100*retained/total

retention_rate
    
```

## convert the returned dictionary into a dataframe

```python
retention_rate_df=pd.DataFrame.from_dict(retention_rate, orient='index')
retention_rate_df=retention_rate_df.rename({0:'retention_rate'},axis=1)
retention_rate_df
```

## Sort the values

```python
retention_rate_df.sort_values('retention_rate',ascending=False)[:20]
```

```python
plt.bar(retention_rate_df.index, retention_rate_df.retention_rate)
plt.title('Customer Retention Rate')
plt.xlabel('State')
plt.xticks(rotation=90)
plt.ylabel('Retention Rate(%)')

plt.show()
```
